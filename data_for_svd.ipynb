```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
marketing_df = pd.read_csv('bank_Marketing_Data.xls')
```

```python
month_mapping = {
    'jan': 1,
    'feb': 2,
    'mar': 3,
    'apr': 4,
    'may': 5,
    'jun': 6,
    'jul': 7,
    'aug': 8,
    'sep': 9,
    'oct': 10,
    'nov': 11,
    'dec': 12
}
list_for_one_hot_encoding = ['marital', 'education', 'contact', 'poutcome', 'job']
```

```python
marketing_df['default'] = marketing_df['default'].apply(lambda x : int(x == 'yes'))
marketing_df['housing'] = marketing_df['housing'].apply(lambda x : int(x == 'yes'))
marketing_df['loan'] = marketing_df['loan'].apply(lambda x : int(x == 'yes'))
marketing_df['month'] = marketing_df['month'].map(month_mapping)
marketing_df['deposit'] = marketing_df['deposit'].map(lambda x: int(x == 'yes'))
```

```python
encoder = OneHotEncoder(sparse=False)
# Применяем OneHotEncoder только к категориальным признакам
encoded_categorical = encoder.fit_transform(marketing_df[list_for_one_hot_encoding])
```

```python
# Получаем названия новых закодированных признаков
encoded_feature_names = encoder.get_feature_names_out(list_for_one_hot_encoding)

# Преобразуем закодированные данные в DataFrame для удобства
encoded_df = pd.DataFrame(encoded_categorical, columns=encoded_feature_names)
# Соединяем обратно с числовыми признаками
numerical_features = marketing_df.drop(columns=list_for_one_hot_encoding)
df_with_one_hot = pd.concat([numerical_features, encoded_df], axis=1)
```
